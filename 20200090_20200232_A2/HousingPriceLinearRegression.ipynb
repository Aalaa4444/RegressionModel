```python
X_train, X_test, y_train, y_test = train_test_split(X_train,y_train, test_size=0.33, random_state=42)
m = len(X_train)
```

```python
# data is stored in numpy array/matrix
print(f"X Shape: {X_train.shape}, X Type:{type(X_train)})")
print(X_train)
print(f"y Shape: {y_train.shape}, y Type:{type(y_train)})")
print(y_train)
```

```python
b_init = 9999999999
w_init=(np.linalg.inv((X_train.T) @ X_train)@(X_train.T))@ y_train
print(f"w_init shape: {w_init.shape}, b_init type: {type(b_init)}")
w_init
```

```python
def loss(X, y, w):    
    m = X.shape[0]
    cost = 0.0
    f_wb = X @ w 
    cost = (1/(2*m)) * np.sum((f_wb-y)**2)                              
    return(np.squeeze(cost))    
```

```python
# Compute and display cost using our pre-chosen optimal parameters. 
cost = loss(X_train, y_train, w_init)
print(f'Cost at optimal w : {cost}')
```

```python
def compute_gradient(X, y, w): 
    m,n = X.shape          
    dj_dw = np.zeros((n,))
    dj_db = 0.
    f_wb = X @ w             
    e   = f_wb - y                
    dj_dw  = (1/m) * (X.T @ e)    
    dj_db  = (1/m) * np.sum(e)    
    return dj_db,dj_dw
```

```python
#Compute and display gradient 
tmp_dj_db, tmp_dj_dw = compute_gradient(X_train, y_train, w_init)
print(f'dj_db at initial w,b: {tmp_dj_db}')
print(f'dj_dw at initial w,b: \n {tmp_dj_dw}')
```

```python
import copy,math
def gradient_descent(X, y, w_in, cost_function, gradient_function, alpha, num_iters): 
    m = len(X)
    hist={}
    hist["cost"] = []; hist["params"] = []; hist["grads"]=[]; hist["iter"]=[];
    w = copy.deepcopy(w_in)     #avoid modifying global w within function
    
    save_interval = np.ceil(num_iters/10000) # prevent resource exhaustion for long runs
    print(f"Iteration Cost          w0       w1       w2       w3       b       djdw0    djdw1    djdw2    djdw3    djdb  ")
    print(f"---------------------|--------|--------|--------|--------|--------|--------|--------|--------|--------|--------|")
    for i in range(num_iters):
        dj_db,dj_dw = gradient_function(X, y, w)   
        w = w - alpha * dj_dw               
        #b = b - alpha * dj_db               
        if i == 0 or i % save_interval == 0:     
            hist["cost"].append(cost_function(X, y, w))
            hist["params"].append([w])
            hist["grads"].append([dj_dw,dj_db])
            hist["iter"].append(i)
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0:
            #print(f"Iteration {i:4d}: Cost {cost_function(X, y, w, b):8.2f}   ")
            cst = cost_function(X, y, w)
            print(f"{i:9d} {cst:0.5e} {w[0]: 0.1e} {w[1]: 0.1e} {w[2]: 0.1e} {dj_dw[0]: 0.1e} {dj_dw[1]: 0.1e} {dj_dw[2]: 0.1e} {dj_db: 0.1e}")
    return w, hist
```

```python

initial_w = np.zeros_like(w_init)
initial_b = 1150000
iterations = 60000
alpha = 1.0e-1
# run gradient descent 
w_final, J_hist = gradient_descent(X_train, y_train, w_init,
                                                    loss, compute_gradient, 
                                                    alpha, iterations)
print(f"b,w found by gradient descent:{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) }, target value: {y_train[i]}")
```
